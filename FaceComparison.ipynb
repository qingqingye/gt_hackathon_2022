```python
from tqdm import tqdm
```

```python
import os
from itertools import combinations
import face_recognition
import numpy as np
import os
import shutil
```

```python
picture_names = os.listdir("output_cacd_has_under18")
file_names = picture_names
picture_names = [picture_name.split(".")[0] for picture_name in picture_names]
picture_names = [picture_name.split("_") for picture_name in picture_names]
```

```python
print(file_names[1])
print(picture_names[1])
```

```python
print(range(len(picture_names)))
```

```python
# {name:{age:[file name], age:[file name, file name], name:{...}}}
celebrity_dic = {}
for i in range(len(picture_names)):
    pic = picture_names[i]
    age = pic[0]
    name = "".join(pic[1:-1])
   
    if name not in celebrity_dic:
        different_age = {}
        different_age[age] = [file_names[i]]
        celebrity_dic[name] = different_age
    else:
        # this cele has in dic 
        if age in celebrity_dic[name]:
            celebrity_dic[name][age].append(file_names[i])
        else:   
            celebrity_dic[name][age] = [file_names[i]]   
```

```python
for name in tqdm(celebrity_dic.keys()):
    destination = "dump"
    source = "output_cacd_has_under18/"
    for age in celebrity_dic[name].keys():
        pic_filenames = celebrity_dic[name][age]

        for pic_filename in pic_filenames:
            path = os.path.join("output_cacd_has_under18", pic_filename)
            image = face_recognition.load_image_file(path)
            image_encoding = face_recognition.face_encodings(image)
            if not image_encoding:
                pic_filenames.remove(pic_filename)
                shutil.move(source+pic_filename,destination)

        combs = list(combinations(pic_filenames,2))
        # Compare combinations of images samilarities
        face_distances = []
        for comb in range(len(combs)):
            image1 = face_recognition.load_image_file(os.path.join("output_cacd_has_under18", combs[comb][0]))
            image2 = face_recognition.load_image_file(os.path.join("output_cacd_has_under18", combs[comb][1]))
            try:
                image1_encoding = face_recognition.face_encodings(image1)[0]
                image2_encoding = face_recognition.face_encodings(image2)[0]
                distances = face_recognition.face_distance([image1_encoding], image2_encoding)
                face_distances.append(distances)
            except:
                print(image1)
                print(image2)
                print(comb,"stoped")
                continue
   
        if all(np.array(face_distances) > 0.65):
            for pic_filename in pic_filenames:
                shutil.move(source+pic_filename,destination)
        else:
            # find min combination and take one
            min_distance_index = face_distances.index(min(face_distances))
            base_pic = combs[min_distance_index][0]
            left_pic_filenames = pic_filenames.remove(base_pic)

            image1 = face_recognition.load_image_file(source+base_pic)
            image1_encoding = face_recognition.face_encodings(image1)[0]

            if left_pic_filenames:
                for left_pic_filename in left_pic_filenames:
                    image2 = face_recognition.load_image_file(left_pic_filename)
                    image2_encoding = face_recognition.face_encodings(image2)[0]

                    face_distances = face_recognition.face_distance([image1_encoding], image2_encoding)

                    if face_distances > 0.65:
                        shutil.move(source+left_pic_filename,destination+pic_filename)
```
